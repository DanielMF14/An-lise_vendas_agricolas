    * A realização deste projeto tem como objetivo por em prática e revisar os conhecimentos de pandas, matplotlib, numpy e seaborn. 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("dados_agricultura.csv")
```

```python
display(df)
```

```python
# Alterando o nome das colunas do nosso DF:
df.columns = ['ID_produto', 'Nome_produto', 'Categoria', 'Preco/KG', 'Unidades_enviadas/KG', 'Unidades_vendidas/KG', 'Unidades_em_estoque/KG', 'Fornecedor', 'Fazenda', 'Data_venda']
```

```python
display(df)
```

```python
# Verificando os tipos de categoria e suas respectivas quantidades:
qtd_categorias = df["Categoria"].value_counts(ascending=False)
```

```python
print(qtd_categorias)
```

```python
# Vamos elaborar um gráfico para melhor visualização:
qtd_categorias.plot(kind='barh')
plt.title('Categorias e suas respectivas quantidades.')
```

* Podemos observar uma distribuição bem variada de categorias, mas ainda sim, as categorias de frutas e laticínios se destacam com uma maior quantidade produtos em suas categorias.

```python
# Vendo a média dos preços por categoria:
media_categorias = df.groupby('Categoria')['Preco/KG'].mean().round(2).sort_values(ascending= False)
```

```python
# Vizualização dos preços em um gráficos:
plt.figure(figsize=(10, 6))
media_categorias.plot(kind='bar', color='skyblue')
plt.title('Média de Preço por Categoria', fontsize=16)
plt.xlabel('Categoria', fontsize=12)
plt.ylabel('Média do Preço (por KG)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

* Nós podemos perceber que os produtos da categoria "gado" possuem um valor médio mais alto em comparação com as demais categorias. 

```python
# Extraindo os 5 produtos mais vendidos em termo de unidades vendidas:
mais_vendidos = df["Nome_produto"].value_counts(ascending=False).head(5)
print(mais_vendidos)
```

```python
plt.figure(figsize=(12, 6))
mais_vendidos.plot(kind='bar', color='green')
plt.title('Produtos mais vendidos')
plt.xlabel('Produtos')
plt.ylabel('Quantidade de vendas')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

* Podemos observar que os produtos tiveram uma quantidade de vendas muito próximas, mas ainda sim o queijo apresenta uma maior quantidade de transações.
* É notável também que todos os produtos são muito comuns na dieta rotineira das pessoas.
* Uma outra análise válida é que dos 5 produtos mais vendidos, 4 são da categoria laticínios e 1 é da categoria de aves.

* Criando gráficos para ver a tendência de vendas por mês e de vendas mensais por categoria.

```python
# Vamos converte a coluna de data para datetime:
df["Data_venda"] = pd.to_datetime(df["Data_venda"])
```

```python
# Vamos criar uma nova coluna, extraindo o ano e o mês da coluna Data_venda:
df['Mes_Ano'] = df["Data_venda"].dt.to_period('M') 
```

```python
# Vamos agrupar por 'Mes_Ano' e somar as unidades vendidas:
vendas_por_mes = df.groupby("Mes_Ano")["Unidades_vendidas/KG"].sum()
```

```python
plt.figure(figsize=(12, 6))
vendas_por_mes.plot(kind='line', marker='o', color='skyblue')
plt.title('Tendência de Vendas ao Longo do Tempo', fontsize=16)
plt.xlabel('Mês/Ano', fontsize=12)
plt.ylabel('Total de Unidades Vendidas (kg)', fontsize=12)
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()
```

* Elaborando gráficos para ver a tendência de vendas por catêgori ao longo de um ano. 

```python
vendas_por_categoria = df.groupby(['Mes_Ano', 'Categoria'])['Unidades_vendidas/KG'].sum().unstack(fill_value=0)
```

```python
num_categorias = vendas_por_categoria.shape[1]  # Número de categorias
fig, axs = plt.subplots(num_categorias, 1, figsize=(12, 4 * num_categorias), sharex=True)

# Plotar cada categoria em um subgráfico
for i, categoria in enumerate(vendas_por_categoria.columns):
    axs[i].plot(vendas_por_categoria.index.astype(str), vendas_por_categoria[categoria], marker='o', label=categoria)
    axs[i].set_title(f'Tendência de Vendas: {categoria}', fontsize=16)
    axs[i].set_ylabel('Total de Unidades Vendidas (kg)', fontsize=12)
    axs[i].legend()
    axs[i].grid()

plt.xlabel('Mês/Ano', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
